```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
import sklearn
```

```python
df=pd.read_csv('movies.csv')
```

```python
df.info()
```

```python
df['genres']=df['genres'].str.split('|',expand=False)
df=df.explode('genres')
```

```python
df.head()
```

```python
cross_df=pd.crosstab(df['title'],df['genres'])
```

```python
cross_df.head()
```

```python
from scipy.spatial.distance import pdist, squareform
```

```python
jaccard_distances = pdist(cross_df.values, metric='jaccard')
```

```python
square_jaccard_distances = squareform(jaccard_distances)
```

```python
jaccard_similarity_array = 1 - square_jaccard_distances
```

```python
distance_df = pd.DataFrame(jaccard_similarity_array,index=cross_df.index,columns=cross_df.index)
```

```python
distance_df.head()
```

```python
def find_match(title_name):
  matches=distance_df[title_name].sort_values(ascending=False)
  return matches.head(20)
```

```python
find_match('\'Hellboy\': The Seeds of Creation (2004)')
```
